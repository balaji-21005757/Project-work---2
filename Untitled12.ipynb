```python
import groq
import os

# Set up Groq API client (Replace with your API key)
api_key = "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"  # Replace with your actual API key
client = groq.Client(api_key=api_key)

# Function to analyze news and classify as True or Fake
def analyze_news(news_text):
    prompt = f"""
    Given the following news article, determine if it is likely to be true or fake.
    Analyze based on facts, credibility, and potential misinformation.

    News Article:
    "{news_text}"

    Respond with 'True' if the news is likely accurate, and 'Fake' if it is misinformation.
    """

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",  # Using Llama 3.1
        messages=[{"role": "user", "content": prompt}],
        temperature=0.2,  # Lower for factual accuracy
        max_tokens=10  # Short response for classification
    )

    return response.choices[0].message.content.strip()

# Example News Articles
news_samples = [
    "NASA confirms water on the surface of Mars.",
    "Aliens found living in underground cities on Earth!",
    "Government launches new AI policy to regulate big tech."
]

# Perform detection and store results
results = {news: analyze_news(news) for news in news_samples}

# Print results
for news, verdict in results.items():
    print(f"News: {news}\nPrediction: {verdict}\n")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np

# Sample dataset: Actual vs Predicted labels
true_labels = ["True", "Fake", "True", "Fake", "Fake", "True", "Fake", "True"]
predicted_labels = ["True", "Fake", "True", "True", "Fake", "True", "Fake", "Fake"]  # Model's predictions

# Convert labels to binary (1 for Fake, 0 for True)
y_true = np.array([1 if label == "Fake" else 0 for label in true_labels])
y_pred = np.array([1 if label == "Fake" else 0 for label in predicted_labels])

# Compute metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)
conf_matrix = confusion_matrix(y_true, y_pred)

# Print metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


news_samples = [
    "NASA confirms water on the surface of Mars.",
    "Aliens found living in underground cities on Earth!",
    "Government launches new AI policy to regulate big tech."
]


predicted_labels = [0, 1, 0]  # Assume AI model's prediction


true_labels = [0, 1, 0]  # True (0), Fake (1)


accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels)
recall = recall_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels)
conf_matrix = confusion_matrix(true_labels, predicted_labels)


for news, label in zip(news_samples, predicted_labels):
    classification = "Fake" if label == 1 else "True"
    print(f"News: {news}\nPrediction: Based on the given news article, I would classify it as '{classification}'.\n")

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["True", "Fake"], yticklabels=["True", "Fake"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]
values = [accuracy, precision, recall, f1]

plt.figure(figsize=(8, 5))
sns.barplot(x=metrics, y=values, palette="viridis")
plt.ylim(0, 1)
plt.ylabel("Score")
plt.title("Performance Metrics")
for i, v in enumerate(values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha="center", fontsize=12)
plt.show()
```
